```python
# initial config
from trainer import Trainer
from models import closest_l2_distance
import tensorflow as tf
import pickle, os
import numpy as np

batch_size = 50
image_size = 224
lr = 0.005
epoch = 10
checkpoint_dir = './data/'
```

```python
# feature extractor ready
demo_module = Trainer(batch_size, image_size, 0., 1)
demo_module.build_model()

demo_module.saver = tf.train.Saver()
ckpt = tf.train.latest_checkpoint(checkpoint_dir)
if ckpt:
    demo_module.saver.restore(demo_module.sess, ckpt)
    print(" [*] Parameter restored from %s"%ckpt)
else:
    print(" [!] Not found checkpoint in (%s), process terminates"%checkpoin_dir)
    raise ValueError

demo_module.sess.run([tf.global_variables_initializer(), tf.local_variables_initializer()])
```

```python
# embedding to semantic space to retrieval
from scipy import misc
% matplotlib inline
import matplotlib.pyplot as plt

if os.path.exists("./data/retrieval.pkl"):
    fin = open('./data/retrieval.pkl', 'rb')
    feature_and_paths = pickle.load(fin)
    fin.close()
else:
    test_files = demo_module.d_loader._d['val_files']

    feature_and_paths = []
    for idx, f in enumerate(test_files):
        print(idx)
        img = misc.imread(f)
        img = misc.imresize(img, (image_size, image_size)) * (1. / 255) - 0.5
        
        feature = demo_module.demo(test_img = img)
        feature_and_paths.append( (feature, f) )
        
        fout = open("./data/retrieval.pkl", "wb")
        pickle.dump(feature_and_paths, fout)
        fout.close()
        
    print(" [*] retrieval data saved at ./data/retrieval.pkl")
print(" [*] retrieval ready")
```

```python
feature = demo_module.demo(test_img = img)

paths = closest_l2_distance(feature_and_paths, feature)

plt.figure(1)
plt.imshow(misc.imread(paths[0]))
plt.figure(2)
plt.imshow(misc.imread(paths[1]))
plt.figure(3)
plt.imshow(misc.imread(paths[2]))
plt.figure(4)
plt.imshow(misc.imread(paths[3]))
plt.figure(5)
plt.imshow(misc.imread(paths[4]))

plt.show()
```
